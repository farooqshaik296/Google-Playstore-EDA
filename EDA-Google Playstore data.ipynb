```python
def size_process(item):
    if str(item)[-1] == 'M':
        res = float(str(item).replace('M',''))  # Initialize res with a value
        res = res * 1024
        return res
    elif str(item)[-1] == 'k':
        res = float(str(item).replace('k',''))  # Initialize res with a value
        return res
    else:
        return np.NaN  # No need to use str() here since NaN is a float
        
        
```

```python
df_copy['Size'] = df_copy['Size'].apply(size_process)
```
